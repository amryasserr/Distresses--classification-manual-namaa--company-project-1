```python
def image_index(x):
    
   num=x.split("+")
   num_1=int(num[0][1:7])
   num_2=float(num[1])
   if num_2<5:
        if len(str(num_1*2+1)) == 1:
            return(f"-Romdas360-0-0000{num_1*2+1}.jpg")
        elif len(str(num_1*2+1)) == 2:
            return(f"-Romdas360-0-000{num_1*2+1}.jpg")
        elif len(str(num_1*2+1)) == 3:
            return(f"-Romdas360-0-00{num_1*2+1}.jpg")
        elif len(str(num_1*2+1)) == 4:
            return(f"-Romdas360-0-0{num_1*2+1}.jpg")
        elif len(str(num_1*2+1)) == 5:
            return(f"-Romdas360-0-{num_1*2+1}.jpg")
   elif num_2>=5:
        if len(str(num_1*2+2)) == 1:
            return(f"-Romdas360-0-0000{num_1*2+2}.jpg")
        elif len(str(num_1*2+2)) == 2:
            return(f"-Romdas360-0-000{num_1*2+2}.jpg")
        elif len(str(num_1*2+2)) == 3:
            return(f"-Romdas360-0-00{num_1*2+2}.jpg")
        elif len(str(num_1*2+2)) == 4:
            return(f"-Romdas360-0-0{num_1*2+2}.jpg")
        elif len(str(num_1*2+2)) == 5:
            return(f"-Romdas360-0-{num_1*2+2}.jpg")
```

```python
LCMS_Crack_Processed["NEW_IMAGE_INDEX"] = Survey_ID + LCMS_Crack_Processed["IMAGE_FILE_INDEX"].apply(image_index)
LCMS_Crack_Processed["NEW_IMAGE_INDEX"]
```
